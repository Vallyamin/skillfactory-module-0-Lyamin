# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


ВАРИАНТ 2

```python
answers['27'] = '(Daniel Radcliffe, Rupert Grint), (Daniel Radcliffe, Emma Watson)'
```

```python
origin = pd.DataFrame(data['original_title'], columns=['original_title'])
casts = pd.DataFrame(data['cast'].str.split(
    '|', expand=True).stack(), columns=['cast'])
result = pd.merge(origin.reset_index(), casts.reset_index(), left_on=[
                  'index'], right_on=['level_0'], how='inner')
xs = []
for film in result.level_0.unique():
    for item in itertools.combinations(result[result.level_0 == film].cast, 2):
        xs.append(item)
pd.Series(xs).value_counts().head(10)
```
