```python
import numpy as np
def game_core_vl(number):
    count=1
    predict = np.random.randint(1,101)
    print(predict)
    while number != predict:
        count+=1
        if number//10 < predict//10:
            predict-=10
        else:
            predict+=10
        if number%10 < predict%10:
            predict-=1
        else:
            predict+=1
        print(predict)
    return(count)
```
